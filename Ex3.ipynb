```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
df = pd.read_csv('C:/Users/Yerassyl/Desktop/data mining/shanghai_ranking_2024.csv')
numeric_df = df.select_dtypes(include=['float64', 'int64'])
scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(numeric_df), columns=numeric_df.columns)
print(df_standardized.head())

print(df.head())
categorical_columns = df.select_dtypes(include=['object']).columns
print(categorical_columns)
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
print(df_encoded.head())

print(df.columns)
print(df.head())
bins = [0, 50, 100, 150, 200, 300]
labels = ['0-50', '51-100', '101-150', '151-200', '201+']
df['Award'] = pd.cut(df['Award'], bins=bins, labels=labels, right=True)
print(df.head())

```
